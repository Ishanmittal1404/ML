#### Prac 3: Apply Back Propagation Algorithm

```python
# initail weights
iw = {0 : 1,
      1 : 1,
      2 : 0,
      3 : 1}
total_nodes = 6
lr = 0.9
final = 6
initial_node = [1,2,3]
```

```python
import pandas as pd 
df = pd.read_csv('data.csv')
```

```python
df
```

```python
li = []
for i in range(1,total_nodes+1):
    if i not in initial_node:
        li.append(i)
li
```

```python
import math 
def solve(df):
    net_input = {}
    output = {}
    error = {}
    
    for i in li:
        count = 0
        if i == final:
            for j in range(len(df['j'])):
                if i == df['j'][j]:
                    if df['i'][j] == 0:
                        count += df['val'][j]
                    else:
                        count += output[int(df['i'][j])]*(df['val'][j])
            net_input[i] = round(count,3)
            output[i] = 1/(1 + math.exp(-1*net_input[i]))

        else:
            for j in range(len(df['j'])):
                if i == df['j'][j]:
                    count += iw[int(df['i'][j])]*(df['val'][j])
            net_input[i] = round(count,1)
            output[i] = 1/(1 + math.exp(-1*net_input[i]))
    for i in li[::-1]:
        if i == final:
            error[i] = output[i]*(1-output[i])**2
        else:
            for j in range(len(df['i'])):
                if df['i'][j] == i and df['j'][j] == 6:
                    val_loc = j
                    break
            error[i] = output[i]*(1-output[i])*error[final]*df['val'][val_loc]
    output.update(iw)
    
    updated_weights = {}
    df1 = df
    for i in range(len(df['i'])):
        val1 = df['val'][i] + (lr*error[int(df['j'][i])]*output[int(df['i'][i])])
        df1['val'][i] = round(val1,3)
    return df1,error
```

```python
count = 0
old_error = {}
while True:    
    df1,error= solve(df)
    df = df1
    print(error)
    count += 1
    if old_error == error:
        break
    else:
        old_error = error

print('\n\n')
print('Total Iterations Required:',count)
print(df)
```
