#### Prac 7:  Apply Naive Bayes Algorithm on a large dataset and get the accuracy for the same

```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv("Dataset.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df, test_size=0.20, random_state=0)
```

```python
list_of_column_names = []
for row in train:
    list_of_column_names.append(row)
```

```python
countY=0
countN=0
dataset=np.array(train[list_of_column_names[-1]])
for i in range(len(dataset)):
    if(dataset[i]== 1):
        countY=countY+1
    elif(dataset[i]== 0 ):
        countN=countN+1
print(countY)
print(countN)
```

```python
def findProbForUsingFeature(index,feature):
    unique_values=train[feature].unique()
    dataset1=np.array(train[list_of_column_names[index]])
    dataset2=np.array(train[list_of_column_names[-1]])
    computations=[]
    for value in unique_values:
        countYes=0
        countNo=0
        for row in range(len(dataset1)):
            if(dataset1[row]==value and dataset2[row]== 1):
                countYes=countYes+1
            elif(dataset1[row]==value and dataset2[row]== 0):
                countNo=countNo+1
        computations.append([feature,value,countYes,countNo,float(countYes)/float(countY),float(countNo)/float(countN)])
    return computations
```

```python
computations0 = findProbForUsingFeature(0,list_of_column_names[0])
computations1 = findProbForUsingFeature(1,list_of_column_names[1])
computations2 = findProbForUsingFeature(2,list_of_column_names[2])
computations3 = findProbForUsingFeature(3,list_of_column_names[3])
computations4 = findProbForUsingFeature(4,list_of_column_names[4])
computations5 = findProbForUsingFeature(5,list_of_column_names[5])
computations6 = findProbForUsingFeature(6,list_of_column_names[6])
computations7 = findProbForUsingFeature(7,list_of_column_names[7])
```

```python
test = test.reset_index()
test
```

```python
final = []
for i in range(len(test)):
    list1 = []
    l1 = test[list_of_column_names[0]][i]
    l2 = test[list_of_column_names[1]][i]
    l3 = test[list_of_column_names[2]][i]
    l4 = test[list_of_column_names[3]][i]
    l5 = test[list_of_column_names[4]][i]
    l6 = test[list_of_column_names[5]][i]
    l7 = test[list_of_column_names[6]][i]
    l8 = test[list_of_column_names[7]][i]
    list1 = [l1,l2,l3,l4,l5,l6,l7,l8]
    final.append(list1)
```

```python
final
```

```python
def yes(today):
    P_Yes_Today=1
    for i in computations0:
        if(i[1]==today[0]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations1:
        if(i[1]==today[1]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations2:
        if(i[1]==today[2]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations3:
        if(i[1]==today[3]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations4:
        if(i[1]==today[4]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations5:
        if(i[1]==today[5]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations6:
        if(i[1]==today[6]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    for i in computations7:
        if(i[1]==today[7]):
            P_Yes_Today=P_Yes_Today*((i[2])/(countY))

    P_Yes_Today=P_Yes_Today*((countY)/len(df))
    return P_Yes_Today

def no(today):
    P_No_Today=1
    for i in computations0:
        if(i[1]==today[0]):
            P_No_Today=P_No_Today*((i[3])/(countN))
        
    for i in computations1:
        if(i[1]==today[1]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    for i in computations2:
        if(i[1]==today[2]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    for i in computations3:
        if(i[1]==today[3]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    for i in computations4:
        if(i[1]==today[4]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    for i in computations5:
        if(i[1]==today[5]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    for i in computations6:
        if(i[1]==today[6]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    for i in computations7:
        if(i[1]==today[7]):
            P_No_Today=P_No_Today*((i[3])/(countN))

    P_No_Today=P_No_Today*((countN)/len(df))
    return P_No_Today
```

```python
output = []
for i in final:
    P_Yes_Today = yes(i)
    P_No_Today = no(i)
    try:
        P_Yes=float(P_Yes_Today)/float(P_Yes_Today+P_No_Today)
        P_No=float(P_No_Today)/float(P_Yes_Today+P_No_Today)
    except:
        P_Yes = 1
        P_No = 0
    if(P_Yes>=P_No):
        output.append(1)
        print("Prediction is 1")
    else:
        output.append(0)
        print("Prediction is 0") 
```

```python
count = 0
for i in range(len(output)):
    if int(output[i]) == int(test['Outcome'][i]):
        count +=1
print('Accuracy: ',count/len(test))
```
