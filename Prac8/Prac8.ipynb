#### Prac 8: Understand and implement SVM rules on the iris data. The rules would be Linear, Polynomial, Sigmoid and Gaussian

```python
from sklearn import svm
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris=load_iris()
print("Features : ",iris.feature_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3,random_state=400)
```

```python
clf_linear = svm.SVC(kernel='linear')
clf_linear.fit(X_train, y_train)

clf_polynomial=svm.SVC(kernel='poly',degree=10)
clf_polynomial.fit(X_train,y_train)

clf_sigmoid=svm.SVC(kernel='sigmoid')
clf_sigmoid.fit(X_train,y_train)

clf_gaussian=svm.SVC(kernel='rbf')
clf_gaussian.fit(X_train,y_train)

y_pred_linear = clf_linear.predict(X_test)
y_pred_polynomial = clf_polynomial.predict(X_test)
y_pred_sigmoid = clf_sigmoid.predict(X_test)
y_pred_gaussian = clf_gaussian.predict(X_test)

print("Accuracy for Linear Kernal:",metrics.accuracy_score(y_test, y_pred_linear))
print("Accuracy for Polynomial Kernal:",metrics.accuracy_score(y_test, y_pred_polynomial))
print("Accuracy for Sigmoid Kernal:",metrics.accuracy_score(y_test, y_pred_sigmoid))
print("Accuracy for Gaussian Kernal:",metrics.accuracy_score(y_test, y_pred_gaussian))
```

```python
X = iris.data[:, :2] 
y = iris.target
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)
clf_polynomial2=svm.SVC(kernel='linear',random_state=True)
clf_polynomial2.fit(X,y)

Z = clf_polynomial2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Linear Kernel')
plt.show()

plt.subplot(1, 1, 1)
clf_polynomial2=svm.SVC(kernel='poly')
clf_polynomial2.fit(X,y)

Z = clf_polynomial2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Polynomial Kernel')
plt.show()

plt.subplot(1, 1, 1)
clf_polynomial2=svm.SVC(kernel='sigmoid',random_state=True)
clf_polynomial2.fit(X,y)

Z = clf_polynomial2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Sigmoid Kernel')
plt.show()

plt.subplot(1, 1, 1)
clf_polynomial2=svm.SVC(kernel='rbf',random_state=True)
clf_polynomial2.fit(X,y)

Z = clf_polynomial2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Gaussian Kernel')
plt.show()
```
